**Replacing Not available and unknown values with nan**

```python
#we need to remove 'not available' and 'unkown' data
data = pd.read_csv('orders.csv', na_values=['Not Available','unknown'])
#we make them Nan values
```

```python
data.head(20)
```

```python
data['Ship Mode'].unique()
```

**Renaming Column Names**

```python
#renaming the columns (method 1)
data.rename(columns={'Order Id':'order_id'}, inplace=True)
```

```python
data.columns
```

```python
#renaming the columns (method 2)
data.columns.str.lower()
```

```python
#now assisgning the string to data.columns
data.columns = data.columns.str.lower()
```

```python
data.columns
```

```python
# now replacing blank values with '-'
data.columns = data.columns.str.replace(' ','_')
```

```python
data.columns
```

```python
data.head()
```

**Deriving New Columns**

```python
#derive new column discount
data['discount'] = data['list_price']*data['discount_percent']*0.01
data.head()
```

```python
#derive new column sale_price
data['sale_price']= data['list_price']-data['discount']
data.head()
```

```python
#deriving profit column
data['profit'] = data['sale_price'] - data['cost_price']
```

```python
data.head()
```

```python
data.dtypes
```

```python
#converting order_data datatype from object(Str) to datetime
data["order_date"]=pd.to_datetime(data["order_date"],format="%Y-%m-%d")
```

```python
data.dtypes
```

**DATA CLEANING: Dropping Redundant column**

```python
data.drop(columns=['list_price','cost_price','discount_percent'])
```

```python
#above command doesnt directly impact our main data
data
```

```python
#using inplace command to drop from original data
data.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
```
